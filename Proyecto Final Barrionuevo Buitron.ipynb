```python
## LECTURA DEL ARCHIVO DE TEXTO QUE CORRESPONDE A LA BASE DE PROFESORES
import numpy as np
import matplotlib.pyplot as plt
#Libreria seaborn para hacer histogramas tambien
import seaborn as sns
datos_profesores = np.loadtxt('Base_datos_evaluacion_1.txt',delimiter='\t',dtype=str)
datos_profesores
```

#### 1. LISTAR LOS EMPLEADOS QUE PERTENCEN A LA FACULTAD DE ADMINISTRACIÓN

```python
## **************************************************************************
## RESPUESTA A LA PREGUNA NRO.1 
## LISTAR LOS EMPLEADOS QUE PERTENCEN A UNA FACULTAD DETERMIANDA: 2
## **************************************************************************

def empleados_x_facultad(empleados, nombre):
    num_filas = range(len(empleados))
    print("******************************************************")
    print("Lista de empleados que pertenecen a la facultad de --> "+ nombre +'\n')
    j=4
    print("    Nombre                 Facultad"+'\n')
    for i in num_filas:
        if empleados[i][4] == nombre:
            print(empleados[i][2] + '   '+ empleados[i][1] + '    -->      ' + empleados[i][4] )



## LISTAR LAS PERSONAS QUE PERTENECEN A LA FACULTAD DE ADMINISTRACION           
empleados_x_facultad(datos_profesores,'Administración')
```

#### 2. LISTAR LOS EMPLEADOS QUE TIENEN EL CARGO DE DOCENTE

```python
## **************************************************************************
## RESPUESTA A LA PREGUNA NRO.2
## LISTAR LOS EMPLEADOS QUE TIENEN UN CARGO DETERMINADO
## **************************************************************************

def empleados_x_cargo(empleados, nombre):
    num_filas = range(len(empleados))
    print("******************************************************")
    print("Lista de empleados que tienen el cargo de --> "+ nombre +'\n')
    j=5
    print("    Nombre                 Cargo"+'\n')
    for i in num_filas:
        if empleados[i][j] == nombre:
            print(empleados[i][2] + '   '+ empleados[i][1] + '    -->      ' + empleados[i][j] )


#LISTAR LOS EMPLEADOS QUE TIENEN EL CARGO DE DOCENTE
empleados_x_cargo(datos_profesores,'Docente')
```

#### 3. SACAR EL TOTAL DE EMPLEADOS QUE ESTÁN ASIGANDOS A LA FACTULTA DE ADMINISTRACIÓN

```python
## **************************************************************************
## RESPUESTA A LA PREGUNA NRO.3
## SACAR EL TOTAL DE EMPLEADOS QUE ESTÁN ASIGANDOS A LA FACTULTAD DE ADMINISTRACIÓN
## **************************************************************************

def total_empleados_x_cargo(empleados, nombre):
    num_filas = range(len(empleados))
    print("******************************************************")
    j=4
    contar_emp = 0
    for i in num_filas:
        if empleados[i][j] == nombre:
            contar_emp = contar_emp + 1
    print('El Área Ádministrativa tiene ' + str(contar_emp) + '  empleados')

## LISTAR EL TOTAL DE EMPLEADOS QUE ESTÁN ASIGNADOS A LA ÁREA ADMINISTRATIVA
total_empleados_x_cargo(datos_profesores,'Administración')
```

#### 4. SACAR EL TOTAL DE EMPLEADOS QUE TIENE EL CARGO DE DOCENTE ADMINISTRATIVO

```python
# **************************************************************************
## RESPUESTA A LA PREGUNA NRO.4
## SACAR EL TOTAL DE EMPLEADOS QUE TIENEN UN CARGO DETERMINADO 
## **************************************************************************

def total_empleados_x_cargo(empleados, nombre):
    num_filas = range(len(empleados))
    print("******************************************************")
    j=5
    contar_emp = 0
    for i in num_filas:
        if empleados[i][j] == nombre:
            contar_emp = contar_emp + 1
    print(' Existen ' + str(contar_emp) + ' empleados con el cargo de "' + nombre+ '"')

## LISTAR EL TOTAL DE EMPLEADOS QUE ESTÁN ASIGNADOS A LA ÁREA ADMINISTRATIVA
total_empleados_x_cargo(datos_profesores,'Docente')
```

#### 5. SACAR EL SUELDO PROMEDIO DE TODO EL PERSONAL


```python
# **************************************************************************
## RESPUESTA A LA PREGUNA NRO.5
## SACAR EL PROMEDIO DE EDAD DE TODO EL PERSONAL 
## **************************************************************************

def sueldo_promedio(empleados):
    num_filas = range(len(empleados))
    contar_emp = 0
    suma_sueldo = 0.0
    sueldo_string =''
    for i in num_filas:
        contar_emp = contar_emp + 1
        sueldo_string = empleados[i][6]
        suma_sueldo =  suma_sueldo + float(sueldo_string.replace('$','').replace('.',''))
    promedio = suma_sueldo / contar_emp
    print("******************************************************")
    print('El sueldo promedio del todo el personal es: $ ' + str(round(promedio,2)) + '')

## SACAR EL SUELDO PROMEDIO DE TODO EL PERSONAL
sueldo_promedio(datos_profesores)
```

#### 6. SACAR LA DESVICACIÓN ESTANDAR DEL SUELDO


```python
# **************************************************************************
## RESPUESTA A LA PREGUNA NRO.5
## SACAR LA DESVICACIÓN ESTANDAR DEL SUELDO (SIGMA)
## ME DICE QUE TAN DISPERSOS SON RESPECTO A UN PROMEDIO
## **************************************************************************
def desviacion_estanda(empleados):
    sueldos = np.asfarray(np.array([s.lstrip("$") for s in empleados[:,6]]))
    desviacion = np.std(sueldos)
    return desviacion
print("******************************************************")
print("La desviacion de los sueldos de los trabajadores es:  " + str(desviacion_estanda(datos_profesores)))


```

#### 7. SACAR LA EDAD MAXIMA Y MINIMA DEL PERSONAL


```python
# **************************************************************************
## RESPUESTA A LA PREGUNA NRO.5
## SACAR LA EDAD MAXIMA Y MINIMA DEL PERSONAL
## **************************************************************************
def sacar_edad_max(empleados):
    return max(empleados[:,9])
def sacar_edad_min(empleados):
    return min(empleados[:,9])
print("*************************************")
print("La Edad de la persona mas longeva es " + str(sacar_edad_max(datos_profesores)))
print("La Edad de la persona menos longeva es " + str(sacar_edad_min(datos_profesores)))
```

#### 8. SACAR EL PROMEDIO DE LA EDAD DEL PERSONAL


```python
def sacar_promedio_edad(empleados):
    return np.mean(np.asfarray(datos_profesores[:,9]))
print("*************************************")
print("El promedio de Edades es " + 
      str(sacar_promedio_edad(datos_profesores)) + 
      " => es decir " + 
      str(round(sacar_promedio_edad(datos_profesores))) + " anios")
```

#### 9. MOSTRAR UN HISTOGRAMA DEL PERSONAL POR FACULTAD.
